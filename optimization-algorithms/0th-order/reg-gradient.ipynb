```python
import numpy as np
import matplotlib.pyplot as plt
 # this is very expensive for high dimensions
def regression_gradient(f, x, m, d):
    fx = f(x)
    n = len(x)
    rd = np.random.randn(m,n)
    dX = d * (rd/np.linalg.norm(rd, axis=1, keepdims=True))
    df = np.array([f(x + dx) - fx for dx in dX])
    return np.linalg.lstsq(dX, df, rcond=None)[0]

def f(x):
    return np.cos(x[0])/np.sin(x[1])

x = np.array([1.5, 2.5])
g = regression_gradient(f, x, m=20, d=1e-5)
print("Approximate gradient: ", g)
x1, x2 = x[0], x[1]
true_gradient = np.array([
    -np.sin(x1) / np.sin(x2),
    -np.cos(x1) * np.cos(x2) / (np.sin(x2) ** 2)
])

print("True gradient:", true_gradient)
                                             
```
