# Combinatorics and Factorials
Computing possibilities if we have a sequence where order matters (about permutations) then the formulas are as follows $$P(n, r) = \frac{n!}{(n - r)!}$$

```python
def factorial(n):
    fac = 1
    for i in range(1, n+1):
        fac *= i 
    return fac
```

```python
def permutations(n, r):
    num = factorial(n)
    den = factorial(n-r)
    return num/den
permutations(5, 3)
```

If order doesn't matter and they are just combinations we have $$C(n, r) = \frac{n!}{r!(n - r)!}$$

```python
def combinations(n, r):
    num = factorial(n)
    den = factorial(r) * factorial(n-r)
    return num/den
combinations(5, 3)
```

### Birthday Problem

```python
import matplotlib.pyplot as plt
import numpy as np

def no_one_shares(n, days):
    n = n+1
    num = factorial(days)
    den = factorial(days - n) * days ** n
    return num/den
probs = []

for p in range(365):
    prob = 1 - no_one_shares(n=p, days=365)
    probs.append(prob)

probas = np.array(probs)
r50 = np.min(np.where(probas >= .50))

plt.plot(range(365), probs)
plt.plot(probas[:(r50+1)], color='r')
plt.xlabel('Days')
plt.ylabel('Probs')
plt.axvline(x=r50, color='r'
            , ls='--', lw=0.75) # threshold
plt.axhline(y=0.5, color='r'
            , ls='--', lw=0.75)
plt.xticks(np.array([1, r50, 50, 100, 200, 365]))
plt.yticks(np.array([0.1, 0.3, 0.5, 0.8, 1]))

f'The probability is {probas[r50]:.3f} with {r50} persons'
```
