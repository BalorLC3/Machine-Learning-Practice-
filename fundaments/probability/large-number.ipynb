# Law of large numbers
As the number of $N$ experiments increases, the sample mean approximates the true mean

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

def coins(N):
    tosses = bernoulli.rvs(0.5, size=N)
    cumulative_heads = np.cumsum(tosses == 1)
    probs = cumulative_heads / np.arange(1, N+1)
    return list(range(1, N+1)), probs

N = 300

plt.figure(figsize=(8,6))
for i in range(6):
    plt.plot(*coins(N))
plt.hlines(y=[0.4, 0.6], xmin=0, xmax=N,
          linestyles='--', color='0.2')
plt.xlim(0, N); plt.ylim(0, 1)
plt.yticks([0.0, 0.2,0.4, 0.5, 0.6, 0.8,1])
plt.xlabel('Number of Tosses')
plt.ylabel('Running proportion of heads')
plt.show()
```
